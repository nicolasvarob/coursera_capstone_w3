```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

## 1.- Create a Dataframe scraping the Wikipedia page

Get the wikipedia page with the list of postal codes of Toronto, Canada. Will store the table with this data in results.

```python
URL = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
results = soup.find("table").find("tbody")
```

Iterate the html response from <b>results</b> to get the variables in a structured form, each row will be stored as a dictionary in an array called <b>tableArr</b>. 

```python
tableArr = []

for td in results.findAll("td"):
    # Will Postal Codes where Borough is not assigned 
    if td.span.text=='Not assigned':
        pass
    else:
        row = {}
        row['PostalCode'] = td.p.text[:3]
        row['Borough'] = (td.span.text).split('(')[0]
        row['Neighborhood'] = (((((td.span.text).split('(')[1]).strip(')')).replace(' /',',')).replace(')',' ')).strip(' ')
        tableArr.append(row)
```

Get the data stored in tableArr into a Dataframe 

```python
df=pd.DataFrame(tableArr)
df.head()
```

```python
print("Dataframe shape is" , df.shape)
```

## 2.- Add geospatial data to each neighborhood

Will use the geospatial coordinate csv as the geocoder package can be unreliable

```python
df_geo = pd.read_csv('Geospatial_Coordinates.csv')
## Rename the postal code column to be joinable with the main dataframe
df_geo = df_geo.rename(columns={'Postal Code': 'PostalCode'})
df_geo.head()
```

Left join the geospatial dataframe to the neighbordhoods dataframe to get both Latitude and Longitude

```python
df_full = df.merge(df_geo,"left","PostalCode")
```

## 3.- Cluster Neighborhoods


```python
# import k-means from clustering stage
from sklearn.cluster import KMeans
import folium # map rendering library
import numpy as np
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values
# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors
```

### Get the coordinates of Toronto so then we can visualize its neighborhoods into a map

```python
address = 'Toronto, CA'

geolocator = Nominatim(user_agent="test")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

### Map visualization of the neighborhoods

```python
# create map of Manhattan using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=11)

# add markers to map
for lat, lng, label in zip(df_full['Latitude'], df_full['Longitude'], df_full['Neighborhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```

###  Neighborhoods clustering
Next, we'll cluster each entry based on their Latitude and Longitude

```python
# set number of clusters
kclusters = 5

grouped_cluster = df_full.drop(['Neighborhood','Borough','PostalCode'], 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(grouped_cluster)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10]
```

Add the column with the respective cluster label we got from Kmeans to our neighborhood dataframe

```python
df_full.insert(0, 'Cluster Labels', kmeans.labels_)
df_full.head()
```

Lastly, visualize the clustered neighborhoods into a map

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(df_full['Latitude'], df_full['Longitude'], df_full['Neighborhood'], df_full['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)

map_clusters
```
